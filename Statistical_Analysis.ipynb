```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings as wr
from scipy.stats import mannwhitneyu
wr.filterwarnings('ignore')
```

```python
df= pd.read_csv('loan.csv',low_memory=False)
```

```python
categorical_col = df.select_dtypes(include='object').columns.to_list()
numeric_col = df.select_dtypes(include='number').columns.to_list()
```

```python
categorical = df.select_dtypes(include='object').columns.tolist()
from scipy.stats import chi2_contingency
for i in categorical:
    loan_crosstab = pd.crosstab(df['loan_condition'], df[i])
    chi, p, dof, expected = chi2_contingency(loan_crosstab)
    significant = p < 0.05
    print(f"Loan_condition and {i} the p-value is", p)
    print("Are the above features associated-------",significant)
    print("==================================================================")
```

```python
df["loan_condition"].value_counts()
```

```python
df_g = df.loc[df['loan_condition'] == 'Good Loan']
df_b = df.loc[df['loan_condition'] == 'Bad Loan']
```

```python
for i in numeric_col:
    t_stat, p_value = stats.ttest_ind(df_g[i],df_b[i])
    alpha = 0.05
    if p_value < alpha:
        print(f"Reject the null hypothesis; there is a significant difference between the {i} for Good and Bad Loan.,{p_value}")
    else:
        print(f"Fail to reject the null hypothesis; there is no significant difference between the between the {i} for Good and Bad Loan,{p_value}")
    print("=======================================================================================================================================")        
```

```python
for i in numeric_col:
    t_stat, p_value = mannwhitneyu(df_g[i],df_b[i])
    alpha = 0.05
    if p_value < alpha:
        print(f"Reject the null hypothesis; there is a significant difference between the {i} for Good and Bad Loan.{p_value}")
    else:
        print(f"Fail to reject the null hypothesis; there is no significant difference between the between the {i} for Good and Bad Loan.{p_value}")
    print("=======================================================================================================================================")        
```
